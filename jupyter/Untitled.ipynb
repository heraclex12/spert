```python
import pandas as pd
file = pd.read_csv('data/ner_artworks.txt')
#

def find_head_idx(source, target):
    target_len = len(target)
    for i in range(len(source)):
        if source[i: i + target_len] == target:
            return i, i + target_len
    return -1
```

```python
data = list()
for row in file.itertuples():
    sentence = row.sentence
    sentence = 
    artwork = row.artwork
    person = row.person
    
    tokenized_artwork = '_'.join(artwork.split(" "))
    tokenize_person = '_'.join(person.split(" "))
    
    tokenized_sentence = sentence.replace(artwork, artwork)
    tokenized_sentence =  tokenized_sentence.replace(person, person)
    
    tokens = tokenized_sentence.split(" ")
    print(tokens)
    find_head_idx(tokens, person.split(" "))
#     artwork_start = tokenized_sentence.find(tokenized_artwork)
#     artwork_end = artwork_start + len(tokenized_artwork) + 1
#     person_start = tokenized_sentence.find(tokenize_person)
#     person_end =  person_start + len(tokenize_person) + 1
    
    artwork_start, artwork_end = find_head_idx(tokens, artwork.split(" "))
    person_start, person_end = find_head_idx(tokens, person.split(" "))
    
    item = dict()
    item.setdefault('tokens', tokens)
    entities = list()
    entities.append({'type': 'artwork', 'start': artwork_start, 'end': artwork_end})
    entities.append({'type': 'person', 'start': person_start, 'end': person_end})
    item.setdefault('entities', entities)
    item.setdefault('relations', [{'type': 'artwork', 'head': 1, 'tail': 0}])
    
    data.append(item)
```

```python
train_data = list()
dev_data = list()
limit = len(data) * 0.85
print(limit)
for d in data:
    if len(train_data) <= limit:
        train_data.append(d)
    else:
        dev_data.append(d)
```
